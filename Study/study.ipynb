```python
import pandas as pd
```

```python
data_df = pd.read_excel('../PhishDataset/data_bal - 20000.xlsx')
data_df.head()
```

```python
study_df = pd.read_excel('phishing ankieta.xlsx')
study_df.head()
```

```python
study_df = study_df.drop(["Sygnatura czasowa", "Wynik", "Ile masz lat"], axis=1, errors='ignore')

def clean_header(col):
    if "[" in col and "]" in col:
        return col.split("[")[1].split("]")[0]
    if col == "Płeć":
        return "sex"
    if col == "Jakie masz wykształcenie":
        return "education"
    if col == "Kierunek studiów, jeśli dotyczy":
        return "study_field"
    return col

study_df.columns = [clean_header(c) for c in study_df.columns]

study_df.head()
```

```python
import pandas as pd
import numpy as np

mapping = {"Phishing": 1, "Bezpieczna strona": 0}

url_cols = [col for col in study_df.columns if "/" in col]

for col in url_cols:
    study_df[col] = study_df[col].map(mapping)

study_df[url_cols] = study_df[url_cols].fillna(study_df[url_cols].mean())
```

```python
def is_it(row):

    study_field = str(row["study_field"]).lower()
    keywords = ["robot","cyberbezpieczeństwo", "informatyka", "computer", "it", "infa", "programowanie", "cs", "teleinf"]
    
    if any(word in study_field for word in keywords):
        return 1
    else:
        return 0

study_df['is_it'] = study_df.apply(is_it, axis=1)

print("kierunki informatyczne")

it_fields = study_df[study_df["is_it"] == 1]["study_field"].unique()

for x in it_fields:
    print(f"- {x}")


print("\nkierunki inne")

it_fields = study_df[study_df["is_it"] == 0]["study_field"].unique()

for x in it_fields:
    print(f"- {x}")
```

```python

survey_urls = [col for col in study_df.columns if "/" in col]

ground_truth = {}

for url in survey_urls:
    match = data_df[data_df['URLs'] == url]
    
    if not match.empty:
        label = int(match.iloc[0]['Labels'])
        ground_truth[url] = label

print(len(ground_truth)/20)
```

```python

y_true = []
y_pred = []

for url, label in ground_truth.items():
    answers = study_df[url].round().astype(int).tolist()
    y_true.extend([label] * len(answers))
    y_pred.extend(answers)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

confusion_matrix = (confusion_matrix(y_true, y_pred))
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score

def get_shots_for_group(df_subset, gt_dict):
    y_true, y_pred = [], []
    url_cols = [col for col in df_subset.columns if col.startswith('http')]
    
    for url in url_cols:
        if url in gt_dict:
            label = gt_dict[url]
            answers = df_subset[url].round().astype(int).tolist()
            y_true.extend([label] * len(answers))
            y_pred.extend(answers)

    return y_true, y_pred


analysis_groups = {
    "Informatycy": {
        "subset": study_df[study_df['is_it'] == 1],
    },
    "Nie-informatycy": {
        "subset": study_df[study_df['is_it'] == 0],
    }
}

def plots(analysis_groups):
    num_groups = len(analysis_groups)
    fig, axes = plt.subplots(1, num_groups, figsize=(9 * num_groups, 7))

    if num_groups == 1: axes = [axes]

    for i, (name, config) in enumerate(analysis_groups.items()):

        y_true, y_pred = get_shots_for_group(config["subset"], ground_truth)
        
        if len(y_true) > 0:

            acc = accuracy_score(y_true, y_pred)
            prec = precision_score(y_true, y_pred)
            rec = recall_score(y_true, y_pred)
            
            cm = confusion_matrix(y_true, y_pred, normalize='true')

            disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Safe', 'Phish'])
            disp.plot(ax=axes[i], values_format='.1%')
            
            axes[i].set_title(
                f"{name} (N={len(y_true)})\n"
                f"accuracy: {acc:.1%} | precision: {prec:.1%}\n"
                f"recall (Czułość): {rec:.1%}", 
                fontsize=12, pad=15
            )
        else:
            axes[i].set_title(f"{name} - Brak danych")

    plt.suptitle("Porównanie skuteczności wykrywania phishingu", fontsize=16, y=1.05)
    plt.tight_layout()
    plt.show()

plots(analysis_groups)
```

```python
analysis_groups_sex = {
    "Kobiety": {
        "subset": study_df[study_df['sex'].str.lower() == 'kobieta'],
    },
    "Mężczyźni": {
        "subset": study_df[study_df['sex'].str.lower() == 'mężczyzna'],
    },
    "Inna": {
        "subset": study_df[study_df['sex'].str.lower() == 'inna'],
    }
}

plots(analysis_groups_sex)
```
